## Are Stocks More Popular If They've Historically Performed Well?

I used the historical changes in stock prices from MarketWatch to calculate the average return for the 50 most popular stocks on Robinhood and then compared these to the return of the S&P 500 over the same time periods. As it turns out, the top 50 most popular stocks outperformed the S&P 500 over each time period, which leads to two possible interpretations.

- Users like to pick stocks that have recently been on the rise.
- The combined opinions of thousands of small investors are able to outperform the S&P 500.
